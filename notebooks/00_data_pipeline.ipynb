```python
async def fetch_pokemon(session, i):
    async with session.get(f'https://pokeapi.co/api/v2/pokemon/{i}') as resp:
        return await resp.json()
```

```python
async with aiohttp.ClientSession() as session:
    tasks = [fetch_pokemon(session, i) for i in range(1, 1024)]
    await tqdm_asyncio.gather(*tasks, desc='fetch pokemon')
```
